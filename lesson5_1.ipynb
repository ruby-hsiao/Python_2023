```python
# 2023-12-03
# 輸入顧客購買金額，若金額在
# 100000元打8折.
# 50000打85折.
# 30000打9折.
# 10000打95折.


import pyinputplus as pyip

discount = 1
while True:
    cost = pyip.inputInt(f"請輸入金額:")
    print(cost)

    if(cost >= 100000):
        cost = cost * 0.8
    elif(cost >= 50000):
        cost  = cost * 0.85
    elif(cost >= 30000):
        cost = cost * 0.9
    elif(cost >= 10000):
        cost  = cost * 0.95

    print(f"購買金額{cost}, 打{discount}折!")

```
